## Changing mutation rate

To start, we'll adjust mu. Because this is a parameter of the underlying simulation, we'll be interfacing with another one of Rosalind's functions, `tregs.simulate.sim()`. 

I'll modify it by hardcoding the static parameter values (everything _other_ than mutation rate) so that it only accepts one argument. This will make it play nice with my first widget, which I'll bind to it.

We should then be able to straightforwardly compute the resulting footprint via a callback every time the widget is updated. Then we simply pass the footprint to the viz code we ported above, and boom!

```python
def sim_update_mutation(mutrate):
    # ugly ass hardcoding for now
    promoter_seq=promoter
    func_pbound=tregs.simulate.simrep_pbound
    binding_site_seqs=[rnap_site, rep_site]
    
    args = [n_NS, n_p, n_r, rnap_emat, rep_emat, ep_wt, er_wt]
    
    preset_mutants=None
    num_mutants=5000
    
    biased=False
    allowed_alph=None
    scaling_factor=100

    #####
    
    if preset_mutants is None:
        if biased:
            mutants = tregs.simulate.biased_mutants(promoter_seq,
                                    mutrate=mutrate,
                                    num_mutants=num_mutants,
                                    allowed_alph=allowed_alph)
        else:
            mutants = np.unique(tregs.simulate.mutations_rand(promoter_seq,
                                        rate=mutrate,
                                        num_mutants=num_mutants,
                                        allowed_alph=allowed_alph,
                                        number_fixed = True,
                                        keep_wildtype=True))
    else:
        mutants = preset_mutants
    
    regions = []
    for bss in binding_site_seqs:
        start, end = tregs.simulate.find_binding_site(promoter_seq, bss)
        regions.append((start, end))

    df_sim = tregs.simulate.sim_helper(mutants, func_pbound, regions, *args)
    
    dna_cnt = tregs.simulate.get_dna_cnt(len(df_sim))
    df_sim['ct_0'] = dna_cnt
    df_sim = df_sim[df_sim.ct_0 != 0.0]

    df_sim['ct_1'] = 0.1 + df_sim['ct_0'] * df_sim['pbound'] * scaling_factor
    df_sim['ct_1'] = df_sim['ct_1'].astype(int)

    df_sim['ct_0'] = df_sim['ct_0'].astype(float)
    df_sim['ct_1'] = df_sim['ct_1'].astype(float)
    df_sim['norm_ct_1'] = df_sim['ct_1'] / df_sim['ct_0']

    return df_sim
```

```python
# define panel widgets (sliders)
mu_slider = pn.widgets.FloatSlider(name="", start=0.03, end=0.3, step=0.01, value=0.03)

# empty plot pane
p_pane = pn.pane.Bokeh()

def update_plot(event):
    # get new df from sim
    footprint = sim_update_mutation(event.new)
    # produce a new plot & update the pane
    p_pane.object = plot_footprint_bokeh(
        promoter,
        footprint,
        region_params,
        x_lims=(-45, 25),
        outfile=None,
        legend_xcoord=1.05
    )

# link slider to update function
mu_slider.param.watch(update_plot, 'value')

# call the update function to initialize
# update_plot(pn.callbacks.Event(mu_slider, new=mu_slider.value))

# arrange dashboard
dashboard = pn.Column(mu_slider, p_pane)

dashboard
```

## Changing library size

In the same way, we'll build upon the previous code by adding a callback for library size as well.

```python
def sim_update(mutrate, num_mutants):
    # ugly ass hardcoding for now
    promoter_seq=promoter
    func_pbound=tregs.simulate.simrep_pbound
    binding_site_seqs=[rnap_site, rep_site]
    
    args = [n_NS, n_p, n_r, rnap_emat, rep_emat, ep_wt, er_wt]
    
    preset_mutants=None
    
    biased=False
    allowed_alph=None
    scaling_factor=100

    #####
    
    if preset_mutants is None:
        if biased:
            mutants = tregs.simulate.biased_mutants(promoter_seq,
                                    mutrate=mutrate,
                                    num_mutants=num_mutants,
                                    allowed_alph=allowed_alph)
        else:
            mutants = np.unique(tregs.simulate.mutations_rand(promoter_seq,
                                        rate=mutrate,
                                        num_mutants=num_mutants,
                                        allowed_alph=allowed_alph,
                                        number_fixed = True,
                                        keep_wildtype=True))
    else:
        mutants = preset_mutants
    
    regions = []
    for bss in binding_site_seqs:
        start, end = tregs.simulate.find_binding_site(promoter_seq, bss)
        regions.append((start, end))

    df_sim = tregs.simulate.sim_helper(mutants, func_pbound, regions, *args)
    
    dna_cnt = tregs.simulate.get_dna_cnt(len(df_sim))
    df_sim['ct_0'] = dna_cnt
    df_sim = df_sim[df_sim.ct_0 != 0.0]

    df_sim['ct_1'] = 0.1 + df_sim['ct_0'] * df_sim['pbound'] * scaling_factor
    df_sim['ct_1'] = df_sim['ct_1'].astype(int)

    df_sim['ct_0'] = df_sim['ct_0'].astype(float)
    df_sim['ct_1'] = df_sim['ct_1'].astype(float)
    df_sim['norm_ct_1'] = df_sim['ct_1'] / df_sim['ct_0']

    return df_sim
```

```python
# define panel widgets (sliders)
mu_slider = pn.widgets.FloatSlider(name="", start=0.03, end=0.3, step=0.01, value=0.03)
n_slider = pn.widgets.FloatSlider(name="library size", start=1, end=10000, step=100, value=5000)

# empty plot pane
p_pane = pn.pane.Bokeh()

def update_plot(event):
    # get new df from sim
    mu = mu_slider.value
    n = n_slider.value

    # get new df from sim
    footprint = sim_update(mutrate=mu, num_mutants=n)
    
    # produce a new plot & update the pane
    p_pane.object = plot_footprint_bokeh(
        promoter,
        footprint,
        region_params,
        x_lims=(-45, 25),
        outfile=None,
        legend_xcoord=1.05
    )

# link slider to update function
mu_slider.param.watch(update_plot, 'value')
n_slider.param.watch(update_plot, 'value')

# call the update function to initialize
# update_plot(pn.callbacks.Event(mu_slider, new=mu_slider.value))

# arrange dashboard
dashboard = pn.Column(mu_slider, n_slider, p_pane)

dashboard
```

Cool. Now, we have a few options for serving this stuff online.

1. Rewrite the backend in JavaScript so that the simulations can run online. Pros: can use a static site like GitHub pages, and avoid the mess of spinning up a server. Cons: big time investment with dubious returns.
2. Bite the bullet and get a Python server going to handle the simulations on the backend, i.e. host a dynamic site. Pros: lots of support via Bokeh and Panel for this kind of thing. Cons: server setup overhead and uncertain scaling -- although I may be able to use existing rpgroup infra?
3. Cheat. Precalculate a bunch of footprints for predefined parameter combinations and save them to the repo, then just do the plotting via BokehJS. I'm pretty sure this is the kind of thing Tom did for his footprints website. Pros: static site! Plus I can probably serialize the footprints quite well with e.g. parquet. Cons: scales horribly with increasing parameter count and resolution, but I don't know the threshold at which this becomes limiting from a performance POV.

Presently the last option seems most attractive to me. So:
- I'll write a wrapper to run the simulations for preset parameter values, serialize them, and save to disk.
- Then I'll modify the dashboard widgets to feed this data into the plotting function, no further simulations required. Widgets will need to be adjusted to match the corresponding parameter space.

```python
def premake_footprints(range_mutrate, range_num_mutants):
    """Run tregs simulations for given parameter ranges, serializing results."""

    print('this will generate', len(range_mutrate)*len(range_num_mutants), 'footprints')

    for mutrate in range_mutrate:
        for num_mutants in range_num_mutants:
            footprint = sim_update(mutrate, num_mutants)

            # TODO: better formatting for long floats
            fname = f'sim-mu{mutrate}-libsize{num_mutants}'
            # print(fname)
            
            footprint.to_parquet(f'data/footprints/{fname}.gzip', compression='gzip')

# np.linspace: start, stop, number; includes end
# np.arange: start, stop, step; half-open interval
mus = np.linspace(0, 0.5, 5)
ns = np.arange(10, 6000, 1000)

# rosalind's sim expects native python floats for mutrate, not np.float64
mus = [float(mu) for mu in mus]

premake_footprints(mus, ns)
```
