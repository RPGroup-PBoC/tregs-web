### Adding interactivity: mutation rate and library size

Because mutation rate and library size are parameters of the underlying simulation, we'll be interfacing with another one of Rosalind's functions, `tregs.simulate.sim()`. 

I'll modify it by hardcoding the static parameter values (everything else) so that it only accepts the needed arguments. This will make it play nice with bound widgets.

We should then be able to straightforwardly compute the resulting dataframe via a callback every time the widgets are updated. Then we simply pass the df to the viz code we ported above, and boom!

First, my reworked simulator:

```python
def sim_update(mutrate, num_mutants):
    # ugly ass hardcoding for now
    promoter_seq=promoter
    func_pbound=tregs.simulate.simrep_pbound
    binding_site_seqs=[rnap_site, rep_site]
    
    args = [n_NS, n_p, n_r, rnap_emat, rep_emat, ep_wt, er_wt]
    
    preset_mutants=None
    
    biased=False
    allowed_alph=None
    scaling_factor=100

    #####
    
    if preset_mutants is None:
        if biased:
            mutants = tregs.simulate.biased_mutants(promoter_seq,
                                    mutrate=mutrate,
                                    num_mutants=num_mutants,
                                    allowed_alph=allowed_alph)
        else:
            mutants = np.unique(tregs.simulate.mutations_rand(promoter_seq,
                                        rate=mutrate,
                                        num_mutants=num_mutants,
                                        allowed_alph=allowed_alph,
                                        number_fixed = True,
                                        keep_wildtype=True))
    else:
        mutants = preset_mutants
    
    regions = []
    for bss in binding_site_seqs:
        start, end = tregs.simulate.find_binding_site(promoter_seq, bss)
        regions.append((start, end))

    df_sim = tregs.simulate.sim_helper(mutants, func_pbound, regions, *args)
    
    dna_cnt = tregs.simulate.get_dna_cnt(len(df_sim))
    df_sim['ct_0'] = dna_cnt
    df_sim = df_sim[df_sim.ct_0 != 0.0]

    df_sim['ct_1'] = 0.1 + df_sim['ct_0'] * df_sim['pbound'] * scaling_factor
    df_sim['ct_1'] = df_sim['ct_1'].astype(int)

    df_sim['ct_0'] = df_sim['ct_0'].astype(float)
    df_sim['ct_1'] = df_sim['ct_1'].astype(float)
    df_sim['norm_ct_1'] = df_sim['ct_1'] / df_sim['ct_0']

    return df_sim
```

Now we assemble a basic dashboard:

```python
# define panel widgets (sliders)
mu_slider = pn.widgets.FloatSlider(name="", start=0.03, end=0.3, step=0.01, value=0.03)
n_slider = pn.widgets.FloatSlider(name="library size", start=1, end=10000, step=100, value=5000)

# empty plot pane
p_pane = pn.pane.Bokeh()

def update_plot(event):
    # get new df from sim
    mu = mu_slider.value
    n = n_slider.value

    # get new df from sim
    df = sim_update(mutrate=mu, num_mutants=n)

    # produce a new plot & update the pane
    p_pane.object = plot_df_bokeh(
        promoter,
        df,
        region_params,
        x_lims=(-45, 25),
        outfile=None,
        legend_xcoord=1.05
    )

# link slider to update function
mu_slider.param.watch(update_plot, 'value')
n_slider.param.watch(update_plot, 'value')

# call the update function to initialize
# update_plot(pn.callbacks.Event(mu_slider, new=mu_slider.value))

# arrange dashboard
dashboard = pn.Column(mu_slider, n_slider, p_pane)

dashboard
```

## Doing things better

One immediate improvement would be to only recalculate everything upon releasing the mouse click (not constantly updating with every slider increment), but I think we can do even better. Let's think about needed next steps -- we have a few options for serving this stuff online:

1. Rewrite the backend in JavaScript so that the simulations can run online. Pros: can use a static site like GitHub pages, and avoid the mess of spinning up a server. Cons: big time investment with dubious returns.
2. Bite the bullet and get a Python server going to handle the simulations on the backend, i.e. host a dynamic site. Pros: lots of support via Bokeh and Panel for this kind of thing. Cons: server setup overhead and uncertain scaling -- although I may be able to use existing rpgroup infra? And will have to confront the issue of repeating costly simulations.
3. Cheat. Precalculate a bunch of footprints for predefined parameter combinations and save them to the repo, then just do the plotting via BokehJS. I'm pretty sure this is the kind of thing Tom did for his footprints website. Pros: static site! Plus I can serialize everything nicely. Cons: scales badly upfront with increasing parameter count.

Presently the last option seems most attractive to me. So:
- I'll write a wrapper to run the simulations for preset parameter values, **do all the calculations upfront**, serialize the results, and save to disk.
- Then I'll modify the dashboard widgets to feed this data into the plotting function, no further simulations required. Widgets will be adjusted to match the corresponding parameter space.

Paired with the aforementioned widget interaction fix, this should improve perfomance drastically.

### Caching footprints upfront

```python
def premake_footprints(range_mutrate, range_num_mutants, nbins, up_scaling_factor, smoothed, windowsize, promoter):
    """Run tregs simulations for given parameter ranges, serializing results."""

    print('this will generate', len(range_mutrate)*len(range_num_mutants), 'footprints')
    print('working...')

    for mutrate in range_mutrate:
        for num_mutants in range_num_mutants:
            df = sim_update(mutrate, num_mutants)

            ## convert df to footprint
            mut_list = df['seq'].values
            mu_data = df['norm_ct_1']
            upper_bound = up_scaling_factor * np.mean(mu_data) 
            
            footprint = tregs.footprint.get_info_footprint(mut_list, mu_data, promoter, nbins, upper_bound,
                                               pseudocount=10**(-6), smoothed=smoothed, windowsize=windowsize)

            # TODO: better formatting for long floats.
            # decimals? ints? idk
            fname = f'sim-mu{mutrate}-libsize{num_mutants}.json'

            # no longer dealing with a df, just a list!
            # so we can serialize to JSON directly
            with open(f'data/footprints/{fname}', 'w') as h:
                footprint = footprint.tolist()
                exshift = tregs.footprint.get_expression_shift(mut_list, mu_data.values, promoter).tolist()

                data = {'footprint':footprint, 'exshift':exshift}
                json.dump(data, h)

    print('done!')
```

## Web deployment

With an MVP in hand, all that's left is to rewrite stuff in JS and push to the static site repo! I'll do that in the `index.html` file.

**Update:** it [seems](http://justinbois.github.io/bootcamp/2021/lessons/l30_javascript_for_bokeh.html) that it isn't necessary to write _all_ the Bokeh code in native JS. Instead, I can use Bokeh via Python to build the layout and plots. So long as I feed in a `ColumnDataSource` and handle the _callbacks_ with JS, I should be able to `bokeh.io.save()` my Python code as a standalone `html` file!

```python
# prep initial x & y values
# we'll use the footprint corresponding to the initial slider parameter values of mu = 0.05 and n = 100
mu = 0.05
n = 100
query = f'sim-mu{mu}-libsize{n}.json'
fpath = f'data/footprints/{query}'

with open(fpath, 'r') as h:
    data = json.load(h)
    footprint = data['footprint']
    exshift = data['exshift']
    shiftcolors=[('#D56C55' if e > 0 else '#738FC1') for e in exshift]
```

```python
# note we use bokeh's native widgets instead of panel's, at Justin's recommendation
mu_slider = bokeh.models.Slider(title="mutation rate", start=0.05, end=0.3, step=0.05, value=0.05)
n_slider = bokeh.models.Slider(title="library size", start=10**2, end=10**4 + 10**2, step=10**3, value=10**2)

# empty plot pane
p_pane = pn.pane.Bokeh()
    
# produce a new plot & update the pane
p, x = plot_footprint_bokeh(
        promoter,
        footprint,
        exshift,
        region_params,
        x_lims=(-45, 25),
        outfile=None,
        legend_xcoord=1.05,
        smoothed=True,
        windowsize=3
    )

p_pane.object = p

p_pane

# with this approach, we don't explicitly write an `update_plot()` function.
# this is because anytime CDS changes, the plot will update automatically
# so we just need to handle passing in x/y.

# access the ColumnDataSource
### could this be improved by precaching the shiftcolors in JSON too? does it matter much??
cds = bokeh.models.ColumnDataSource(dict(x=x, footprint=footprint, exshift=exshift, shiftcolors=shiftcolors))

p.vbar(source=cds, x="x", width=.8, line_width=0, bottom=0, top="footprint", color="shiftcolors");

# Laying out the widgets is the same
widgets = bokeh.layouts.column(
    bokeh.layouts.Spacer(height=30),
    mu_slider,
    bokeh.layouts.Spacer(height=15),
    n_slider,
    width=200,
)

# Lay out with the plot and widgets
layout = bokeh.layouts.row(p, bokeh.layouts.Spacer(width=15), widgets)

### JAVASCRIPT ZONE (barf)
# quirks: need to pass URLs, not local filepaths, for successful fetching
jscode = """
// widget values
var mu = mu_slider.value;
var n = n_slider.value;
var url = `https://raw.githubusercontent.com/RPGroup-PBoC/tregs-web/main/data/footprints/sim-mu${mu}-libsize${n}.json`

// CDS values
var x = cds.data['x'];
var footprint = cds.data['footprint'];
var exshift = cds.data['exshift'];
var shiftcolors = cds.data['shiftcolors'];

fetch(url)
.then(response => response.json())
.then(data => {
    cds.data = {
        x: x,
        footprint: data.footprint,
        exshift: data.exshift,
        shiftcolors: exshift.map(e => e > 0 ? '#D56C55' : '#738FC1')
    };
        
    cds.change.emit();
})
.catch(error => console.error('Error fetching data:', error));
"""

args = dict(cds=cds, mu_slider=mu_slider, n_slider=n_slider)

# Connect the callback to the sliders
mu_slider.js_on_change("value", bokeh.models.CustomJS(code=jscode, args=args))
n_slider.js_on_change("value", bokeh.models.CustomJS(code=jscode, args=args))

bokeh.io.show(layout)
```

```python
def plot_footprint_bokeh(promoter, df, region_params,
                   nbins=2, up_scaling_factor=1,
                   x_lims=(-45, 25), fig_width=750, fig_height=225, legend_xcoord=1.2,
                   max_signal=None,
                   outfile=None, annotate_stn=True,
                   return_fp=False,
                   smoothed=True, windowsize=3,):

    mut_list = df['seq'].values
    mu_data = df['norm_ct_1']
    upper_bound = up_scaling_factor * np.mean(mu_data)

    footprint = tregs.footprint.get_info_footprint(mut_list, mu_data, promoter, nbins, upper_bound,
                                               pseudocount=10**(-6),
                                               smoothed=smoothed, windowsize=windowsize)
    exshift_list = tregs.footprint.get_expression_shift(mut_list, mu_data.values, promoter)

    if max_signal is None:
        max_signal = max(footprint)
        
    total_signal = 0

    p = bokeh.plotting.figure(
        title="test title",
        x_axis_label="Position relative to TSS",
        y_axis_label="Information (bits)",
        width=fig_width,
        height=fig_height,
        aspect_ratio=3.33,
        y_range=(0, max_signal*1.15),
        x_range=(x_lims[0], x_lims[1])
    )

    if smoothed:
        cut = int((windowsize - 1) / 2)
        x = np.arange(-115 + cut, 45 - cut)
    else:
        x = np.arange(-115, 45)
    shiftcolors = [('#D56C55' if exshift > 0 else '#738FC1') for exshift in exshift_list]
    
    # add the footprint
    p.vbar(x, width=.8, line_width=0, bottom=0, top=footprint, color=shiftcolors)
    
    return p
```
